```python
import openmc 
import numpy as np
import matplotlib.pyplot as plt
```

```python
colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown', 'tab:pink', 'tab:gray', 'tab:olive', 'tab:cyan']

fig = openmc.plot_xs('U235', data_type='nuclide', types=['total', 'fission'])
ax1 = fig.gca()

openmc.plot_xs('U238', data_type='nuclide', types=['total', 'fission'], axis=ax1)

lines = fig.gca().get_lines()

for lineIdx, line in enumerate(lines):
    line.set_color(colors[lineIdx//2])
    if (lineIdx % 2 == 1): line.set_linestyle('-.')

plt.xlabel('$E_n$ [eV]')
plt.ylabel('$\sigma$ [b]')
plt.grid()
plt.legend(['U-235 total', 'U-235 fission', 'U-238 total', 'U-238 fission'])
plt.title('Total and fission cross sections for U-235 and U-238')
plt.tight_layout()

image_to_thesis.save(plt, image_to_thesis.ThesisSection.THEORY, "U235-U238-fission-total-cross-section.pdf")

plt.show()

```
